```python
import os
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import tkinter as tk
from tkinter import messagebox, ttk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from datetime import datetime
import importlib.util
from importlib.util import spec_from_file_location, module_from_spec
```

# Initial API call to get and clean the stocks data (EXECUTE ONLY TO CREATE NEW DATA FILES)
for i in range(2014,2022):
    ## Load the list of S&P 500 tickers
    file_path = 'sp_500_stocks.csv'  # Ensure the file is in the same folder as the notebook
    sp500_stocks = pd.read_csv(file_path)
    tickers = sp500_stocks['Ticker'].tolist()
    
    ## Select the first n tickers
    tickers = tickers
    
    ## Define the output directory
    output_dir = f'sp500_data_{i}'  # Relative path, will create 'sp500_data_2022' in current directory
    os.makedirs(output_dir, exist_ok=True)
    
    ## Define the date range
    start_date = f"{i}-01-01"
    end_date = f"{i+1}-01-01"
    
    ## Fetch and save data
    for ticker in tickers:
        try:
            print(f"Fetching data for {ticker}...")
            data = yf.download(ticker, start=start_date, end=end_date)
            file_path = os.path.join(output_dir, f"{ticker}.csv")
            data.to_csv(file_path)  # Save to CSV
            print(f"Saved {ticker} data to {file_path}")
        except Exception as e:
            print(f"Failed to fetch data for {ticker}: {e}")
    
    ## List the saved files
    print(f"Saved files: {os.listdir(output_dir)}")
    
    ## Delete files smaller than 2 Ko
    size_threshold = 2 * 1024  # 2 Ko in bytes (2 Ko = 2 * 1024 bytes)
    for file_name in os.listdir(output_dir):
        file_path = os.path.join(output_dir, file_name)
        if os.path.isfile(file_path):
            file_size = os.path.getsize(file_path)
            if file_size < size_threshold:
                print(f"Deleting {file_name} (Size: {file_size / 1024:.2f} Ko)")
                os.remove(file_path)
    
    print(f"Remaining files after cleanup: {len(os.listdir(output_dir))}")

for year in range(2014,2022):

            output_file_path = os.path.join(output_directory, filename)
    
            with open(input_file_path, "r") as infile:
                lines = infile.readlines()
    
            # Replace the first line and remove the second line
            lines[0] = "Date,Price,Adj Close,Close,High,Low,Open,Volume\n"
            lines.pop(2)
            lines.pop(1)
    
            # Write the cleaned content to the output file
            with open(output_file_path, "w") as outfile:
                outfile.writelines(lines)
    
            print(f"Processed and saved: {output_file_path}")

# Function Definitions

```python
# Function to dynamically load a strategy
def load_strategy(strategy_file):
    if not os.path.exists(strategy_file):
        raise FileNotFoundError(f"Strategy file not found: {strategy_file}")

    # Load the module dynamically
    spec = importlib.util.spec_from_file_location("strategy_module", strategy_file)
    strategy_module = importlib.util.module_from_spec(spec)
    spec.loader.exec_module(strategy_module)

    # Ensure the 'strategy' function exists in the module
    if not hasattr(strategy_module, "strategy"):
        raise AttributeError(f"No 'strategy' function found in {strategy_file}")

    return strategy_module.strategy


def simulate_trading_one_stock(file_path, allocation, strategy_file):
    try:
        # Load stock data
        data = pd.read_csv(file_path, parse_dates=["Date"], index_col="Date")
        ticker = os.path.basename(file_path).split('.')[0]

        # Load the strategy function
        strategy_func = load_strategy(strategy_file)

        # Call the strategy to get the final state
        final_cash, final_holdings, trade_log = strategy_func(data, ticker, stock_holdings=0, cash=allocation)

        # Calculate the final portfolio value
        if final_holdings > 0:
            final_price = data['Adj Close'].iloc[-1]  # Use the last available price
            final_value = final_cash + (final_holdings * final_price)
        else:
            final_value = final_cash
        
        return final_value, trade_log

    except Exception as e:
        print(f"Error processing {file_path}: {e}")
        return allocation, []  # Return initial allocation in case of error
    
    
# Main simulation logic
def main_simulation(strategy_file,total_initial_value, year):
    # Get the list of stock files
    file_name = f'sp500_data_clean_{year}'
    data_dir = os.path.join(os.getcwd(), "sp500_data", file_name)
    stock_files = [file for file in os.listdir(data_dir) if file.endswith('.csv')]
    num_stocks = len(stock_files)
    allocation_per_stock = total_initial_value / num_stocks  # Equal allocation per stock

    # Portfolio tracking
    final_portfolio_value = 0  # Sum of final cash + stock values across all stocks
    all_trade_logs = []  # Consolidated trade log

    # Process each stock file
    for file_name in stock_files:
        file_path = os.path.join(data_dir, file_name)
        stock_final_value, stock_trade_log = simulate_trading_one_stock(
            file_path, allocation_per_stock, strategy_file
        )
        final_portfolio_value += stock_final_value
        all_trade_logs.extend(stock_trade_log)

    # Output results
    print(f"Initial Portfolio Value: ${total_initial_value:.2f}")
    print(f"Final Portfolio Value: ${final_portfolio_value:.2f}")
    return total_initial_value, final_portfolio_value, all_trade_logs
```

```python
# Function to save the trade log to a file
def save_trade_log(trade_log, year, strategy_name):
    # Create logs directory if it doesn't exist
    logs_dir = os.path.join(os.getcwd(), "logs")
    if not os.path.exists(logs_dir):
        os.makedirs(logs_dir)

    # Generate a filename based on strategy and year
    log_filename = f"{strategy_name.split('.')[0]}_{year}_log.txt"
    log_filepath = os.path.join(logs_dir, log_filename)

    # Write trades to the log file
    with open(log_filepath, "w") as log_file:
        log_file.write("Date, Stock, Action, Quantity, Price\n")
        for trade in trade_log:
            log_file.write(f"{trade[0]}, {trade[1]}, {trade[2]}, {trade[3]}, {trade[4]:.2f}\n")

    return log_filepath


# Function to calculate performance metrics
def calculate_metrics(trade_log, initial_value, final_value):
    total_trades = len(trade_log)
    gains = []
    losses = []

    for trade in trade_log:
        if trade[2] == 'SELL':  # On sell, calculate profit or loss
            buy_trade = next(
                t for t in trade_log if t[1] == trade[1] and t[2] == 'BUY' and t[0] < trade[0]
            )
            pnl = (trade[3] * trade[4]) - (buy_trade[3] * buy_trade[4])  # Profit or loss
            if pnl > 0:
                gains.append(pnl)
            else:
                losses.append(abs(pnl))

    cumulative_returns = (final_value - initial_value) / initial_value
    win_ratio = len(gains) / max(len(gains) + len(losses), 1)
    avg_profit_loss = (
        (sum(gains) / len(gains)) if gains else 0
    ) - ((sum(losses) / len(losses)) if losses else 0)

    return cumulative_returns, win_ratio, total_trades, avg_profit_loss


# Function to plot portfolio value
def plot_portfolio_value_in_tkinter(trade_log, initial_cash, year, parent_frame, figsize=(8, 4), dpi=100, padding=(1, 1)):
    cash = initial_cash
    holdings = {}
    portfolio_values = []
    dates = []

    trade_log = sorted(trade_log, key=lambda x: x[0])

    for trade in trade_log:
        date, stock, action, quantity, price = trade
        if action == 'BUY':
            cash -= quantity * price
            holdings[stock] = holdings.get(stock, 0) + quantity
        elif action == 'SELL':
            cash += quantity * price
            holdings[stock] -= quantity
            if holdings[stock] == 0:
                del holdings[stock]

        stock_value = sum(holdings[s] * price for s in holdings)
        portfolio_values.append(cash + stock_value)
        dates.append(date)

    date_range = pd.date_range(start=f"{year}-01-01", end=f"{year+1}-01-01", freq='D')
    portfolio_df = pd.DataFrame(index=date_range, columns=["Portfolio Value"])
    portfolio_df["Portfolio Value"] = None

    for i, date in enumerate(dates):
        portfolio_df.loc[date, "Portfolio Value"] = portfolio_values[i]

    portfolio_df["Portfolio Value"] = portfolio_df["Portfolio Value"].fillna(method='ffill')

    # Create the matplotlib figure
    figure = plt.Figure(figsize=figsize, dpi=dpi)
    ax = figure.add_subplot(111)
    ax.plot(portfolio_df.index, portfolio_df["Portfolio Value"], label='Portfolio Value', color="#4CAF50")
    ax.set_title('Portfolio Value Over Time', fontsize=14, fontweight='bold')
    ax.set_xlabel('Date', fontsize=12)
    ax.set_ylabel('Portfolio Value ($)', fontsize=12)
    ax.legend()
    ax.grid()

    for widget in parent_frame.winfo_children():
        widget.destroy()

    canvas = FigureCanvasTkAgg(figure, master=parent_frame)
    canvas.draw()
    canvas_widget = canvas.get_tk_widget()
    canvas_widget.pack(fill="both", expand=True, padx=padding[0], pady=padding[1])


# Function to refresh strategy files
def refresh_strategy_files(strategy_combobox):
    strategy_folder = "strategies"
    if not os.path.exists(strategy_folder):
        os.makedirs(strategy_folder)
    strategy_files = [f for f in os.listdir(strategy_folder) if f.endswith('.py')]
    strategy_combobox["values"] = strategy_files
    if strategy_files:
        strategy_combobox.set(strategy_files[0])
```
